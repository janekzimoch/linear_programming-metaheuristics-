# Distribution fitting

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import statistics as stat
import pylab
import statsmodels.api as sm
```

```python
x=[80.2, 78.3, 77.8, 76.4, 75.5, 75.4, 75.4, 75.4, 75.3, 75.2, 74.5, 73.9, 73.7, 73.4, 73.3, 72.7, 72.6, 72.5, 72.4, 72.1, 71.9, 71.8, 71.8, 71.7, 71, 70.8, 70.5, 70.3, 70.1, 70.1, 70.1, 70, 70, 69.8, 69.4, 69.3, 69.3, 69.3, 69.2, 69, 68.9, 68.9, 68.7, 68.6, 68.2, 68.2, 68, 67.8, 67.7, 67.1, 67.1, 67.1, 67.1, 66.8, 66.8, 65.8, 65.7, 65.7, 65.6, 65.6, 65.6, 65.3, 65, 64.7, 64.7, 64.5, 64.4, 64.2, 64.2, 63.8, 63.7, 63.3, 63.1, 63.1, 63.1, 62.8, 62.8, 62.3, 62.2, 61.8, 61.8, 61.4, 61.3, 60.7, 60.2, 58.6, 58.6, 57.4, 56.6, 53.5]

plt.hist(x, density=True, bins=20)
plt.ylabel('Historical Occurence');


mu = stat.mean(x)
variance = stat.pvariance(x)
sigma = math.sqrt(variance)
y = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(y, stats.norm.pdf(y, mu, sigma),color='r')

plt.savefig('hist.pdf', transparent=True)
```

```python
stats.probplot(x,dist="norm",plot=pylab)
pylab.savefig("qqplot.pdf", transparent=True)
```

```python
plt.plot(np.sort(x), np.linspace(0, 1, len(x), endpoint=False))

mu=stat.mean(x)
sigma=stat.pstdev(x)

y = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(y, stats.norm.cdf(y, mu, sigma),color='r')

plt.savefig("ecdf.pdf",transparent=True)
```
